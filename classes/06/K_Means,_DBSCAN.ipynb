```python
from sklearn.cluster import KMeans, DBSCAN

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = sns.load_dataset('iris')
df_petal = df.iloc[:, 2:4]
df_petal.head()
```

```python
for i in range(5):
    km = KMeans(n_clusters=3, n_init=1, max_iter=i+1, random_state=235, init='random')
    km.fit(df_petal)

    sns.scatterplot(x='petal_length', y='petal_width', data=df_petal, hue=km.predict(df_petal), palette=['green', 'blue', 'red'])

    plt.scatter(km.cluster_centers_[0, 0], km.cluster_centers_[0, 1], color='darkgreen', marker='d', s=100)
    plt.scatter(km.cluster_centers_[1, 0], km.cluster_centers_[1, 1], color='darkblue', marker='d', s=100)
    plt.scatter(km.cluster_centers_[2, 0], km.cluster_centers_[2, 1], color='darkred', marker='d', s=100)
    plt.title('Iteration #{0}'.format(i+1))
    plt.show()
```

```python
km = KMeans(n_clusters=2, n_init=1, max_iter=100, random_state=235, init='random')
km.fit(df_petal)

sns.scatterplot(x='petal_length', y='petal_width', data=df_petal, hue=km.predict(df_petal))

plt.scatter(km.cluster_centers_[0, 0], km.cluster_centers_[0, 1], color='darkgreen', marker='d', s=100)
plt.scatter(km.cluster_centers_[1, 0], km.cluster_centers_[1, 1], color='darkblue', marker='d', s=100)

plt.show()
```

```python
km = KMeans(n_clusters=5, n_init=1, max_iter=100, random_state=235, init='random')
km.fit(df_petal)

sns.scatterplot(x='petal_length', y='petal_width', data=df_petal, hue=km.predict(df_petal))

plt.scatter(km.cluster_centers_[0, 0], km.cluster_centers_[0, 1], color='darkgreen', marker='d', s=100)
plt.scatter(km.cluster_centers_[1, 0], km.cluster_centers_[1, 1], color='darkblue', marker='d', s=100)
plt.scatter(km.cluster_centers_[2, 0], km.cluster_centers_[2, 1], color='darkred', marker='d', s=100)
plt.scatter(km.cluster_centers_[3, 0], km.cluster_centers_[3, 1], color='orange', marker='d', s=100)
plt.scatter(km.cluster_centers_[4, 0], km.cluster_centers_[4, 1], color='black', marker='d', s=100)
plt.show()
```

```python
for i in [5, 7, 10, 30]:
    dbscan = DBSCAN(min_samples=i)

    dbscan.fit(df_petal)

    sns.scatterplot(x='petal_length', y='petal_width', data=df_petal, hue=dbscan.labels_)
    plt.title('Min samples: {0}'.format(i))
    plt.show()
```

```python
for i in [2, 3, 4, 14]:
    dbscan = DBSCAN(eps=i/10)

    dbscan.fit(df_petal)

    sns.scatterplot(x='petal_length', y='petal_width', data=df_petal, hue=dbscan.labels_)
    plt.title('Epslilon = {0}'.format(i/10))
    plt.show()
```

#  2D Visualization using PCA from scratch

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img = mpimg.imread('PCA.gif')
plt.figure(figsize = (10, 20))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
labels = l.head(15000)
data = d.head(15000)
print("the shape of sample data = ", data.shape)
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

### Find the co-variance matrix, which is: A^T * A



```python
sample_data = standardized_data

# matrix multiplication using numpy
covar_matrix = np.matmul(sample_data.T , sample_data)
print ( "The shape of variance matrix = ", covar_matrix.shape)
```

### Finding the top two eigen-values and corresponding eigen-vectors for projecting onto a 2-Dim space.

```python
sum(covar_matrix[300])
```

```python
from scipy.linalg import eigh

values, vectors = eigh(covar_matrix, eigvals=(782,783))
print("Shape of eigen vectors = ",vectors.shape)
vectors = vectors.T
print("Updated shape of eigen vectors = ",vectors.shape)
```

```python
values
```

```python
# projecting the original data sample on the plane
# formed by two principal eigen vectors by vector-vector multiplication.

import matplotlib.pyplot as plt
new_coordinates = np.matmul(vectors, sample_data.T)
new_coordinates.shape
```

```python
import pandas as pd

# appending label to the 2d projected data
new_coordinates = np.vstack((new_coordinates, labels)).T

# creating a new data frame for ploting the labeled points.
dataframe = pd.DataFrame(data=new_coordinates, columns=( '1st component', '2nd component', "label"))
print(dataframe.head())
```

```python
# ploting the 2d data points with seaborn
import seaborn as sn
sn.FacetGrid(dataframe, hue="label", height=10).map(plt.scatter, '1st component', '2nd component').add_legend()
plt.show()
```

# PCA using Scikit-Learn

```python
# initializing the pca
from sklearn import decomposition
pca = decomposition.PCA()
# configuring the parameteres
# the number of components = 2
pca.n_components = 2
pca_data = pca.fit_transform(sample_data)
# pca_reduced will contain the 2-d projects of simple data
pca_data.shape
# attaching the label for each 2-d data point
pca_data = np.vstack((pca_data.T, labels)).T
# creating a new data fram which help us in ploting the result data
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal", "label"))
sn.FacetGrid(pca_df, hue="label", height=10).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```

# Variance of the data and reduction of the dimensions

```python
# initializing the pca
from sklearn import decomposition
pca = decomposition.PCA()
```

```python
pca.n_components = 784
pca_data = pca.fit_transform(sample_data)

percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_);

cum_var_explained = np.cumsum(percentage_var_explained)

# Plot the PCA spectrum
plt.figure(figsize=(10, 8))
#plt.clf()
plt.plot(cum_var_explained, linewidth=4)
plt.axis('tight')
plt.grid()
plt.xlabel('n_components', size=15)
plt.ylabel('Cumulative_explained_variance', size=15)
plt.show()

# If we take 2-dimensions, approx. 14% of variance is expalined.
# If we take 200-dimensions, approx. 90% of variance is expalined.
print("2D - ", cum_var_explained[2])
print("200D - ",cum_var_explained[200])
```

# t-SNE using Scikit-Learn

```python
from sklearn.manifold import TSNE

data = standardized_data[0:42000,:]
labels = labels[0:42000]

model = TSNE(n_components=2, random_state=0)

# the number of components = 2
# default perplexity = 30
# default learning rate = 200
# default Maximum number of iterations for the optimization = 1000

tsne_data = model.fit_transform(data)

# Ploting the result
tsne_data = np.vstack((tsne_data.T, labels)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Component_1", "Component_2", "label"))
sn.FacetGrid(tsne_df, hue="label", height=6).map(plt.scatter, "Component_1", "Component_2").add_legend()
plt.show()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=5)
tsne_data = model.fit_transform(data)

# creating a new data fram which help us in ploting the result data
tsne_data = np.vstack((tsne_data.T, labels)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))

# Ploting the result of tsne
sn.FacetGrid(tsne_df, hue="label", height=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.title('With perplexity = 5')
plt.show()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=5)
tsne_data = model.fit_transform(data[:1000])

# creating a new data fram which help us in ploting the result data
tsne_data = np.vstack((tsne_data.T, labels[:1000])).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))

# Ploting the result of tsne
sn.FacetGrid(tsne_df, hue="label", height=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.title('With perplexity = 5 (N=1000)')
plt.show()
```
